```python
import pandas as pd
```

```python
df = pd.read_csv('forest_area.csv')
```

```python
df.head()
```

```python
df = df.mask(df == '-')
df = df.fillna(0)
```

```python
df['Total'] = df['Total'].apply(lambda x: float(x.split()[0].replace(',', '')))
```

```python
df['Total'] = df['Total'].astype('float')
```

```python
columns = ['Red pine', 'Eastern white pine',
       'Balsam fir', 'White spruce', 'Black spruce', 'Tamarack',
       'Northern white-cedar', 'Eastern redcedar', 'Other softwoods', 'Oak',
       'Northern hardwoods', 'Lowland hardwoods', 'Cottonwood / Willow',
       'Aspen', 'Birch', 'Balsam poplar']
```

```python
for column in columns:
    df[column] = df[column].astype('float')
```

```python
df.head()
```

```python
for column in columns:
    df[column] = (df[column] / df['Total']) 
```
